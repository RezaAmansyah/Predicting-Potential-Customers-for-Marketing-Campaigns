**Metrik Evaluasi**

Model akan dievaluasi menggunakan ROC AUC Score, dengan fokus utama pada keseimbangan antara recall dan precision, agar model tidak melewatkan pelanggan potensial (FN) namun tetap efisien dari sisi biaya (FP).

model akan menghasilkan dua kemungkinan : 
* prediksi 1 : model memprediksi customer yang menerima penawaran
* prediksi 0 : model memprediski customer yang tidak menerima penawaran 

fokus utama adalah memprediksi customer yang menerima penawaran dari *marketing campaign*

> Hasil Model Prediksi (Confussion Matrix)

| AKtual(Data Asli) | Prediksi Model        | Confusion Matrix | Arti                                                      |
| -------------------- | --------------------- | ------------------------ | --------------------------------------------------------- |
| 1 (Menerima)         | 1 (Prediksi menerima) | **True Positive (TP)**   | Model benar: customer memang menerima                    |
| 0 (Tidak menerima)   | 0 (Prediksi tidak)    | **True Negative (TN)**   | Model benar: customer memang tidak menerima              |
| 0 (Tidak menerima)   | 1 (Prediksi menerima) | **False Positive (FP)**  | Model salah: mengira customer menerima padahal tidak      |
| 1 (Menerima)         | 0 (Prediksi tidak)    | **False Negative (FN)**  | Model salah: melewatkan customer yang sebenarnya menerima|


Untuk mempermudah pemahaman mengenai dampak dari False Positive dan False Negative, mari kita gunakan pendekatan kuantitatif dengan perhitungan numerik sehingga konsekuensi masing-masing dapat terlihat secara lebih jelas dan terukur.

Misalkan:

* Biaya kampanye per pelanggan = $C$ 
* Potensi keuntungan jika pelanggan menerima penawaran = $P$ 
* Jumlah FP =  $N_{FP}$ 
* Jumlah FN =  $N_{FN}$ 

Maka:

$$
\text{Kerugian akibat FP} = N_{FP} \times C
$$

$$
\text{Kerugian akibat FN} = N_{FN} \times P
$$

$$
\text{Total konsekuensi} = (N_{FP} \times C) + (N_{FN} \times P)
$$


Contoh Perhitungan Misal:

* Biaya kampanye per pelanggan  $C$ = 50.000  IDR
* Potensi keuntungan jika pelanggan menerima penawaran $P$ = 200.000  IDR
* Dari model prediksi:

  * FP = 100 pelanggan
  * FN = 50 pelanggan

Maka:

$$
\text{Kerugian FP} = 100 \times 50.000 = 5.000.000 \text{ IDR}
$$

$$
\text{Kerugian FN} = 50 \times 200.000 = 10.000.000 \text{ IDR}
$$

$$
\text{Total konsekuensi} = 5.000.000 + 10.000.000 = 15.000.000 \text{ IDR}
$$



Interpretasi

* FP → biaya kampanye terbuang.
* FN → potensi keuntungan hilang.
* Dengan angka ini, perusahaan bisa **memprioritaskan strategi marketing**: misal lebih hati-hati agar FN rendah, karena kerugiannya lebih besar daripada FP.
* Maka metrik Evaluasi yang akan di gunakan adalah **ROC AUC**

Untuk memahami distribusi data secara lebih mendalam, langkah awal dilakukan identifikasi outlier pada fitur numerik.
Beberapa fitur menunjukkan nilai yang ekstrem, sehingga pada analisis selanjutnya akan digunakan median sebagai ukuran pemusatan. Pendekatan ini dipilih karena median lebih tahan terhadap outlier dibandingkan mean, yang dapat terdistorsi oleh nilai ekstrem.

```python
def calculate_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

outlier_data = []

numerical = [var for var in df.columns if df[var].dtypes != 'object']

for column in numerical:
    outlier_result, lower_bound, upper_bound = calculate_outliers(df, column)
    total_outlier = len(outlier_result)
    outlier_percentage = round(total_outlier / len(df[column]) * 100, 2)
    outlier_data.append([column, total_outlier, outlier_percentage, lower_bound, upper_bound])

outlier_df = pd.DataFrame(outlier_data, columns=["Column", "Total Outliers", "Percentage (%)", "Lower Bound", "Upper Bound"])
outlier_df
```

```python
# Fungsi untuk anotasi jumlah
def annotate_bar_plot(ax):
    for bar in ax.patches:
        ax.annotate(f'{bar.get_height():.0f}',  
                    (bar.get_x() + bar.get_width() / 2., bar.get_height()),  
                    ha='center', va='center',  
                    xytext=(0, -7),  
                    textcoords='offset points',  
                    fontsize=9, color='white', fontweight='semibold')

# Fungsi untuk anotasi persentase
def annotate_bar_plot_percent(ax, total):
    for bar in ax.patches:
        height = bar.get_height()
        ax.annotate(f'{height/total:.2%}',  
                    (bar.get_x() + bar.get_width() / 2., height),  
                    ha='center', va='center',  
                    xytext=(0, 12),  
                    textcoords='offset points',  
                    fontsize=9, color='black', fontweight='semibold')

# Fungsi utama untuk plot semua kolom kategorikal
def plot_categorical_distribution(categorical, ncols=2, palette="viridis"):
    obj_cols = categorical.select_dtypes(include="object").columns.tolist()
    total = len(categorical)
    
    nrows = (len(obj_cols) + ncols - 1) // ncols  # hitung jumlah baris
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(8*ncols, 4*nrows))
    axes = axes.flatten()

    for i, col in enumerate(obj_cols):
        ax = axes[i]
        sns.countplot(data=categorical, x=col, ax=ax, palette=palette, order=categorical[col].value_counts().index,hue=categorical[col])
        ax.set_title(f"Distribusi {col}")
        ax.tick_params(axis="x", rotation=45)

        # Tambahkan anotasi
        annotate_bar_plot(ax)
        annotate_bar_plot_percent(ax, total)

    # Kosongkan axis yang tidak terpakai
    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.style.use('ggplot')
    plt.show()
```

**Check and Handling Outlier Data**

```python
def calculate_outliers_iqr(data, column):
    # Hitung Q1, Q3 dan IQR
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Tentukan batas bawah dan atas
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Cari outlier
    lower_outliers = data[data[column] < lower_bound]
    upper_outliers = data[data[column] > upper_bound]
    
    total_outliers = len(lower_outliers) + len(upper_outliers)
    
    return {
        "lower_bound": lower_bound,
        "upper_bound": upper_bound,
        "total_outliers": total_outliers,
        "lower_outliers_count": len(lower_outliers),
        "upper_outliers_count": len(upper_outliers)
    }

# Simpan hasil
outlier_data = []

numerical = [var for var in df.columns if df[var].dtypes != 'object']

for column in numerical:
    result = calculate_outliers_iqr(df, column)
    outlier_percentage = round(result["total_outliers"] / len(df[column]) * 100, 2)
    
    outlier_data.append([
        column, 
        result["total_outliers"], 
        outlier_percentage, 
        result["lower_bound"], 
        result["upper_bound"], 
        result["lower_outliers_count"], 
        result["upper_outliers_count"]
    ])

outlier_df = pd.DataFrame(outlier_data, columns=[
    "Column", "Total Outliers", "Percentage (%)", 
    "Lower Bound", "Upper Bound", 
    "Lower Outliers", "Upper Outliers"
])

outlier_df
```

**Check Multicolinearity**

tahapan ini akan melihat apakah terdapat kolinearitas antara fitur independen. walaupun kolinearitas tidak mempengaruhi hasil prediksi secara langsung tapi sangat mempengaruhu interpretasi mode,karena akan sulit untuk mengetahui mana fitur yang paling mempengaruhi model. 

pada tahapan ini akan di cek menggunakan VIF (Variance Inflaction Factor). Secara Sederhana VIF adalah alat untuk mengukur tingkat kolinearitas/multikolinearitas. semakin tinggi nilai VIF maka semakin besar ia di jelaskann oleh variabel lain. 



 **Rumus Variance Inflation Factor (VIF)**

Untuk setiap variabel independen    $( X_i )$

$$
VIF_i = \frac{1}{1 - R_i^2}
$$

di mana:

* $( R_i^2 )$= nilai R-squared hasil dari regresi $( X_i)$ terhadap semua variabel independen lainnya.

**Interpretasi**:

| Nilai VIF | Tingkat Kolinearitas     | Interpretasi Praktis                       |
| --------- | ------------------------ | ------------------------------------------ |
| 1         | Tidak ada kolinearitas   | Variabel bebas sepenuhnya independen       |
| 1–5       | Kolinearitas moderat     | Masih bisa diterima                        |
| >5        | Kolinearitas tinggi      | Perlu perhatian                            |
| >10       | Multikolinearitas serius | Sebaiknya variabel dihapus atau diperbaiki |

* N. A. M. R. Senaviratna & T. M. J. A. Cooray. (2019). Diagnosing Multicollinearity of Logistic Regression Model. Asian Journal of Probability and Statistics, 5(2), 1–9. https://doi.org/10.9734/ajpas/2019/v5i230132
* Correlation, Variance Inflation and Multicollinearity in Regression Model. Journal of the Eastern Asia Society for Transportation Studies, 12, 2006–2015. https://doi.org/10.11175/easts.12.2006

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
num = [i for i in df.select_dtypes(include='number').columns]
VIF = [variance_inflation_factor(df[num],i) for i in range(len(df[num].columns))]

vif = pd.DataFrame()
vif['feature'] = df[num].columns
vif['VIF'] = [variance_inflation_factor(df[num],i) for i in range(len(df[num].columns))]
vif
```

**Check Cardinality**


**Cardinality** merupakan jumlah nilai unik yang terdapat pada suatu fitur independen. Tingkat cardinality memiliki pengaruh signifikan terhadap proses **preprocessing data** dan pemilihan **metode encoding** yang digunakan. Apabila sebuah fitur memiliki cardinality yang tinggi, penggunaan metode seperti *One-Hot Encoding* dapat menyebabkan peningkatan kompleksitas data yang berpotensi membuat model menjadi **overfit** serta menurunkan efisiensi proses pelatihan.



```python
def check_cardinality(data):
    cardinality_info = []

    for feature in data.select_dtypes(include='object').columns:
        cardinality = len(data[feature].unique())
        warning = "High" if cardinality > 10 else "Low"
        cardinality_info.append({'Feature': feature, 'Cardinality': cardinality, 'Warning': warning})

    return pd.DataFrame(cardinality_info)

# Contoh penggunaan
cardinality_df = check_cardinality(df)
cardinality_df
```

***Scaling***

**Scaling** merupakan proses penyesuaian skala pada fitur numerik agar seluruh variabel berada dalam rentang yang sebanding. Tujuan utama dari tahap ini adalah memastikan bahwa setiap fitur memiliki **kontribusi yang proporsional** terhadap pembelajaran model machine learning. Tanpa proses scaling, fitur dengan nilai besar—seperti *Income* dalam satuan juta—dapat mendominasi fitur dengan nilai kecil, seperti *Recency* dalam satuan hari. Akibatnya, model dapat menjadi **bias terhadap fitur tertentu** dan menghasilkan performa yang kurang optimal.

Pada penelitian ini, fitur numerik melalui proses **scaling menggunakan RobustScaler**. Metode ini dipilih karena model yang digunakan, seperti **Support Vector Machine (SVM)**, sensitif terhadap perbedaan skala antar fitur. Berbeda dengan metode lain seperti *StandardScaler* atau *MinMaxScaler*, **RobustScaler** memiliki keunggulan dalam **mengurangi pengaruh outlier**, karena proses skalanya didasarkan pada nilai median dan *interquartile range (IQR)*, bukan mean dan standar deviasi.

Cara kerja RobustScaler yaitu dengan **mengurangi nilai median** dari setiap data, kemudian **membaginya dengan Interquartile Range (IQR)** — yaitu selisih antara kuartil ke-75 (Q3) dan kuartil ke-25 (Q1).

Rumusnya:

$$
X_{scaled} = \frac{X - \text{Median}(X)}{IQR}
$$




```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()


pipe = Pipeline([
    ('encoding',ct),
    ('scaler',scaler)
])

pipe
```

1. **Model Benchmarking** — Menguji dan membandingkan performa berbagai algoritma klasifikasi untuk menentukan model dengan hasil terbaik berdasarkan metrik evaluasi utama, yaitu ***ROC AUC Score***. Tahap ini bertujuan untuk mengidentifikasi model yang memiliki kemampuan paling baik dalam membedakan antara kelas positif dan negatif.

2. **Hyperparameter Tuning** — Setelah tiga model dengan performa terbaik teridentifikasi, dilakukan proses *hyperparameter tuning* guna mengoptimalkan parameter masing-masing model. Tujuannya adalah untuk meningkatkan kinerja model secara signifikan dengan menyesuaikan konfigurasi parameter agar sesuai dengan karakteristik data.

3. **Cross Validation** — Proses pelatihan model menggunakan teknik *Stratified K-Fold Cross Validation*, yaitu metode validasi yang membagi data ke dalam beberapa lipatan (fold) dengan proporsi kelas yang seimbang. Tujuan penggunaan teknik ini adalah untuk **mengurangi risiko overfitting**, **memastikan model lebih stabil dan general**, serta **memberikan estimasi performa yang lebih akurat** terhadap data yang belum pernah dilihat sebelumnya.

4. **Randomized Search Optimization** — Dalam proses *hyperparameter tuning*, digunakan metode ***RandomizedSearchCV*** untuk mencari kombinasi parameter terbaik secara acak dalam ruang pencarian yang telah ditentukan. Pendekatan ini dipilih karena lebih **efisien secara waktu dan komputasi** dibandingkan *Grid Search*, namun tetap mampu menghasilkan konfigurasi parameter yang optimal untuk meningkatkan performa model.

```python
# define model 
from sklearn.svm import SVC


logreg = LogisticRegression()
tree = DecisionTreeClassifier(random_state=12,max_depth=5)
knn = KNeighborsClassifier()
svc = SVC()
rf = RandomForestClassifier(random_state=12)
gb = GradientBoostingClassifier(random_state=12)
ada = AdaBoostClassifier(random_state=12)
xgb = XGBClassifier(random_state=12)
cat = CatBoostClassifier(random_state=12)
lgbm = LGBMClassifier(random_state=12,silent=True)

# cross validation 
skf = StratifiedKFold(n_splits=5,shuffle=True, random_state=12)

models = [logreg,tree,knn,svc,rf,gb,ada,xgb,cat,lgbm]
```

Pada tahap **benchmarking**, dilakukan evaluasi dan perbandingan kinerja terhadap sepuluh model klasifikasi dengan menerapkan tiga pendekatan berbeda dalam menangani *imbalanced data*, yaitu:

1. **Tanpa resampling**, sebagai baseline untuk melihat performa model pada data asli.
2. **SMOTE (Synthetic Minority Over-sampling Technique)**, yang menyeimbangkan data dengan menghasilkan sampel sintetis pada kelas minoritas.
3. **SMOTENC (SMOTE for Nominal and Continuous variables)**, yaitu varian SMOTE yang dirancang khusus untuk menangani data campuran antara fitur kategorikal dan numerik.


```python
from imblearn.pipeline import Pipeline


# store result
list_mean_no_resampling = []
list_std_no_resampling = [] 
list_mean_smote = [] 
list_std_smote = [] 
list_mean_smotenc = [] 
list_std_smotenc = [] 

model_name = []

# modeling 

for model in models: 

    # model no resampling
    pipeline_no_resampling = Pipeline([
        ('preprocessing',ct),
        ('scaler',scaler),
        ('model',model)
    ])

    # crossval no resampling
    model_no_resampling=cross_val_score(
        estimator=pipeline_no_resampling,
        X=X_train,
        y=y_train,
        cv = skf,
        scoring='roc_auc',
        n_jobs=-1
    )
    model_name.append(model.__class__.__name__)
    list_mean_no_resampling.append(model_no_resampling.mean())
    list_std_no_resampling.append(model_no_resampling.std())


    # model with smote 

    pipeline_smote=Pipeline([
        ('preprocessing',ct),
        ('scaler',scaler),
        ('smote',smote),
        ('model',model)
    ])

    # crossval 
    model_with_smote=cross_val_score(
        estimator=pipeline_smote,
        X = X_train,
        y = y_train,
        cv = skf,
        scoring='roc_auc',
        n_jobs=-1
    )

    list_mean_smote.append(model_with_smote.mean())
    list_std_smote.append(model_with_smote.std())

    # model with smotenc 

    pipeline_smote=Pipeline([
        ('preprocessing',ct),
        ('scaler',scaler),
        ('smote',smotenc),
        ('model',model)
    ])

    # crossval 
    model_with_smotenc=cross_val_score(
        estimator=pipeline_smote,
        X = X_train,
        y = y_train,
        cv = skf,
        scoring='roc_auc',
        n_jobs=-1
    )

    list_mean_smotenc.append(model_with_smotenc.mean())
    list_std_smotenc.append(model_with_smotenc.std())



```

```python
#hasil cross validasi
pd.DataFrame({
    'Model' : model_name,
    'mean_roc_auc_no_resampling' : list_mean_no_resampling,
    'std_roc_auc_no_resampling' : list_std_no_resampling,
    'mean_roc_auc_smotenc' : list_mean_smotenc,
    'std_roc_auc_smotenc' : list_std_smotenc, 
    'mean_roc_auc_smote' : list_mean_smote,
    'std_roc_auc_smote' : list_std_smote, 
}).sort_values(by='mean_roc_auc_smotenc',ascending=False)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

roc_auc_no_resampling = []
roc_auc_smotenc = []
roc_auc_smote = []

test_model_name = []

for model in models : 

    # define no resampling 
    pipeline_no_resampling = Pipeline([
        ('preprocessing',ct),
        ('scaler',scaler),
        ('model',model)
    ])

    # fit 
    pipeline_no_resampling.fit(X_train,y_train)

    # predict
    y_pred_no_resampling = pipeline_no_resampling.predict(X_test)

    # evaluate model 
    score_no_resampling = roc_auc_score(y_test,y_pred_no_resampling)
    
    # define smotenc 
    pipeline_smotenc = Pipeline([
        ('preprocessing',ct),
        ('scaler',scaler),
        ('smotenc',smotenc),
        ('model',model)
    ])

    # fit 
    pipeline_smotenc.fit(X_train,y_train)

    # predict
    y_pred_smotenc = pipeline_smotenc.predict(X_test)

    # evaluate model 
    score_smotenc = roc_auc_score(y_test,y_pred_smotenc)
    
    # define smote 
    pipeline_smote = Pipeline([
        ('preprocessing',ct),
        ('scaler',scaler),
        ('smote',smote),
        ('model',model)
    ])

    # fit 
    pipeline_smote.fit(X_train,y_train)

    # predict
    y_pred_smote = pipeline_smote.predict(X_test)

    # evaluate model 
    score_smote = roc_auc_score(y_test,y_pred_smote)


    roc_auc_no_resampling.append(score_no_resampling)
    roc_auc_smotenc.append(score_smotenc)
    roc_auc_smote.append(score_smote)
    test_model_name.append(model.__class__.__name__)

score_validation_test = pd.DataFrame({
    'Model' : test_model_name,
    'validation_score_roc_auc_no_resampling' : list_mean_no_resampling,
    'test_score_roc_auc_no_resampling' : roc_auc_no_resampling,
    'validation_score_roc_auc_smotenc' : list_mean_smotenc,
    'test_score_roc_auc_smotenc': roc_auc_smotenc, 
    'validation_score_roc_auc_smote' : list_mean_smote,
    'test_score_roc_auc_smote': roc_auc_smote 
}).sort_values(by='test_score_roc_auc_smotenc',ascending=False)

score_validation_test
```

### **Model Evaluation**



Pada tahap ini dilakukan evaluasi mendalam terhadap performa model untuk memastikan kualitas prediksi yang dihasilkan serta tingkat kemampuan generalisasi model. Proses evaluasi mencakup beberapa langkah berikut:

<p align='center'><b><span style="font-size:22px">SUPPORT VECTOR CLASSIFIER</span></b></p>

**SVC (Support Vector Classifier)** adalah algoritma machine learning yang digunakan untuk mengklasifikasikan data ke dalam dua atau lebih kelompok. SVC bekerja dengan mencari garis batas terbaik (disebut hyperplane) yang memisahkan dua kelas data sejelas mungkin.
Tujuannya: supaya ketika ada data baru, model bisa menebak kelasnya dengan benar.

**Cara Kerja SVC** 


1. **Data dan Label**

   * Diberikan data latih $((x_i, y_i))$ dengan $(x_i)$ adalah fitur, $(y_i)$ adalah label kelas (misal ±1 atau 0/1).
   * Jika kelasnya 0/1, sering diubah ke ±1 untuk formulasi matematika SVM.

2. **Hyperplane & Margin**

   * SVC mencoba mencari sebuah *hyperplane* (garis di 2D, bidang di 3D, hiperruang di dimensi lebih tinggi) yang memisahkan kelas −1 dan +1.

   * Hyperplane ini memiliki bentuk (w^T x + b = 0), di mana:

     * (w) adalah vektor bobot (normal ke hyperplane),
     * (b) adalah bias (jarak hyperplane dari origin secara paralel ke (w)).

   * Model ingin **memaksimalkan margin**, yaitu jarak antara hyperplane dengan *support vectors*—data-data terdekat dari masing-masing kelas ke hyperplane. Margin lebih besar → generalisasi lebih baik.

3. **Soft Margin vs Hard Margin**

   * **Hard Margin**: mengasumsikan kelas-kelas bisa dipisahkan sempurna, sehingga tidak ada kesalahan (tidak ada data di salah sisi hyperplane).
   * **Soft Margin**: memperbolehkan beberapa data melanggar margin / bahkan salah klasifikasi, untuk menangani data yang tidak sepenuhnya separable. Ada trade-off antara memperbesar margin dan mengurangi kesalahan; kontrol lewat parameter **C** (regularization).

4. **Kernel Trick untuk Data Non-Linear**

   * Jika data tidak bisa dipisahkan linier, SVM bisa menggunakan *kernel* — fungsi yang secara implisit memetakan data ke ruang dimensi tinggi, di mana data bisa menjadi linier-separable.
   * Kernel populer: RBF (Gaussian), Polynomial, Sigmoid, Linear (tidak memetakan ke ruang baru).
   * Kernel trick memungkinkan menghitung inner product di ruang tinggi tanpa benar-benar menghitung koordinat di ruang tinggi (menghemat komputasi).

5. **Fungsi Objective (Optimasi)**

   * Formulasi matematisnya adalah meminimalkan:
     $$
     \frac{1}{2} |w|^2 + C \sum_{i=1}^n \xi_i
     $$
     di mana $( \xi_i )$ adalah *slack variables* (berapa banyak margin dilanggar untuk tiap data), dan (C) mengatur seberapa besar penalti terhadap pelanggaran margin/kesalahan.
   * Subyek juga constraint:
     $$
     y_i (w^T x_i + b) \ge 1 - \xi_i, \quad \xi_i \ge 0
     $$

6. **Optimasi (Dual Problem)**

   * Untuk efisiensi dan agar bisa menggunakan kernel, biasanya masalah dipecahkan dalam bentuk *dual* menggunakan Lagrange multipliers.
   * Solusi optimasi menghasilkan bobot untuk tiap data (terutama untuk support vectors). Data yang bukan support vector biasanya tidak mempengaruhi posisi hyperplane.

7. **Prediksi**

   * Setelah model dilatih, untuk prediksi label dari data baru (x), dihitung:
    $$
    \hat{y} = \mathrm{sign}(w^T x + b)
    $$

   * Jika memakai kernel, dihitung sesuai kernelnya menggunakan support vectors dan koefisiennya.

```python

# define final model SVC 
final_model_svc = Pipeline([
    ('preprocessing',ct),
    ('scaler',scaler),
    ('smotenc',smotenc),
    ('model',SVC(
        kernel='rbf',
        gamma='auto',
        class_weight=None,
        C=1,
        probability=True
    ))
])


# fit model
final_model_svc.fit(X_train,y_train)

#predict model
y_pred_final_model_svc = final_model_svc.predict(X_test)

final_score = roc_auc_score(y_test,y_pred_final_model_svc)

pd.DataFrame({
    'Model' : ['SVC'],
    'ROC AUC SCORE' : [final_score]
})
```

***Features Importances***


**Permutation Importance** merupakan metode interpretasi model yang digunakan untuk **mengukur seberapa besar pengaruh setiap fitur terhadap performa model**. Prinsip dasarnya adalah dengan **mengacak (mem–permute)** nilai satu fitur pada dataset uji, kemudian mengamati **penurunan performa model** (misalnya pada metrik ROC AUC).

Jika setelah pengacakan nilai suatu fitur performa model menurun secara signifikan, maka fitur tersebut dianggap **memiliki pengaruh yang besar** terhadap prediksi model. Sebaliknya, jika perubahan performa relatif kecil, maka fitur tersebut memiliki **kontribusi yang rendah**.

Dalam proyek ini, metode **Permutation Importance** digunakan karena **model terbaik yang terpilih adalah Support Vector Classifier (SVC)**.
Model seperti SVC **tidak memiliki atribut internal** untuk menilai pentingnya fitur (seperti *feature_importances_* pada model pohon keputusan) dan **tidak menghasilkan fungsi prediksi yang mudah diinterpretasikan secara langsung**, sehingga metode berbasis *explainability* seperti **SHAP** atau **LIME** menjadi **lebih sulit diterapkan atau memerlukan waktu komputasi yang tinggi**.

Oleh karena itu, **Permutation Importance** dipilih karena:

* Dapat digunakan untuk **model apapun**, termasuk SVC.
* Memberikan **estimasi sederhana namun kuat** tentang kontribusi setiap fitur terhadap performa model.
* **Tidak bergantung pada struktur internal model**, sehingga lebih fleksibel dan model-agnostic.



```python
from sklearn.inspection import permutation_importance

r = permutation_importance(
    estimator=final_model_svc,
    X=X_test,
    y=y_test,
    scoring='roc_auc',
    n_repeats=10,
    random_state=12,
    n_jobs=1
)

feature_names = X_test.columns 

# Membuat DataFrame hasil
df_importance = pd.DataFrame({
    'Feature': feature_names,
    'Importance_Mean': r.importances_mean,
    'Importance_STD': r.importances_std
})

# Urutkan berdasarkan rata-rata Importance
df_importance = df_importance.sort_values(by='Importance_Mean', ascending=False)
display(df_importance)
```
